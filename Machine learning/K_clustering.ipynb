```python
x = data.iloc[:,[3,4]].values
```

```python
x
```

```python
wcss = []
for i in range(1,11):
  kmean = KMeans(n_clusters=i, init = "k-means++",random_state= 42)
  kmean.fit(x)
  wcss.append(kmean.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title("Elbow points graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
kmean = KMeans(n_clusters=3,init = "k-means++",random_state = 0)
y = kmean.fit_predict(x)
y
```

```python
plt.figure(figsize = (8,8))
plt.scatter(x[y==0,0],x[y==0,1],s = 50 ,c ="lightpink",label = "setosa")
plt.scatter(x[y==1,0],x[y==1,1],s = 50 ,c ="blue",label ="virginica")
plt.scatter(x[y==2,0],x[y==2,1],s = 50 ,c ="cyan",label ="versicolor")
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s = 100,c= "r",label = "centroid")
plt.title("k cluster")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
